```python
import numpy as np
import matplotlib.pyplot as plt

# Define function to calculate Secrecy Outage Probability (SOP) for MIMO
def calculate_SOP_MIMO(transmission_power_dB, M, N, threshold=1e-3):
    # Example calculation, replace with your actual SOP calculation function for MIMO
    # For demonstration, let's assume a simple linear relationship between transmission power and SOP
    SOP = 0.1 * np.exp(transmission_power_dB / (M + N - 1) / 10)
    # Ensure SOP values are bounded between 0 and 1
    SOP = np.minimum(np.maximum(SOP, 0), 1)
    return SOP

# Generate transmission power values in dB
transmission_power_dB = np.arange(0, 30, 1)

# Number of transmit antennas (M) and receive antennas (N)
M_values = [1, 2, 3]
N_values = [1, 2, 3]

plt.figure(figsize=(10, 6))

# Plot SOP vs. Transmission Power for different combinations of M and N
for M in M_values:
    for N in N_values:
        SOP_MIMO = calculate_SOP_MIMO(transmission_power_dB, M, N)
        label = f'MIMO (M={M}, N={N})'
        plt.plot(transmission_power_dB, SOP_MIMO, marker='o', label=label)

# Add labels and title
plt.xlabel('Transmission Power (dB)')
plt.ylabel('Secrecy Outage Probability (SOP)')
plt.title('Secrecy Outage Probability (SOP) vs. Transmission Power for MIMO')
plt.grid(True)
plt.legend()
plt.ylim(0, 1)  # Ensure y-axis limits between 0 and 1
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define function to calculate BER for MIMO systems
def calculate_BER_MIMO(SNR, M, N):
    # Example BER calculation for MIMO systems, replace with your actual BER calculation function
    # Here we assume a simple exponential relationship between SNR and BER
    BER = 0.5 * np.exp(-SNR / (M + N - 1))
    return BER

# Generate SNR values in dB
SNR_dB = np.arange(0, 20, 1)

# Convert SNR to linear scale
SNR = 10**(SNR_dB / 10)

# Number of transmit antennas (M) and receive antennas (N) configurations
M_values = [1, 2, 3]
N_values = [1, 2, 3]

plt.figure(figsize=(10, 6))

# Plot BER vs. SNR for different MIMO configurations
for M in M_values:
    for N in N_values:
        BER_MIMO = calculate_BER_MIMO(SNR, M, N)
        label = f'MIMO (M={M}, N={N})'
        plt.semilogy(SNR_dB, BER_MIMO, marker='o', label=label)

# Add labels and title
plt.xlabel('Signal-to-Noise Ratio (SNR) [dB]')
plt.ylabel('Bit Error Rate (BER)')
plt.title('Bit Error Rate (BER) vs. Signal-to-Noise Ratio (SNR) for Various MIMO Configurations')
plt.grid(True)
plt.legend()
plt.show()
```
